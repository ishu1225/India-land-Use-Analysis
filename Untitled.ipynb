```python
# Cleaning the dataset 

import pandas as pd
import numpy as np


df = pd.read_csv('data.csv')  



numeric_cols = [
    'Reported land area ', 
    'Forest land area', 
    'Net sown land area', 
    'Cropped land area', 
    'Land area sown more than once', 
    'Barren and unculturable land area', 
    'Culturable waste land area'
]
print("\nUsing these numeric columns for analysis:", numeric_cols)

# Ensure numeric consistency
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
print("\nInitial Data Types:\n", df.dtypes)

# Check initial missing values
print("\nInitial Missing Values:\n", df.isnull().sum())

# Replace zeros with NaN
df[numeric_cols] = df[numeric_cols].replace(0, np.nan)
print("\nMissing Values After Replacing Zeros:\n", df.isnull().sum())

# Fill missing values with column means
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print("\nMissing Values After Imputation:\n", df.isnull().sum())

# Assign cleaned DataFrame
df_clean = df
print(f"\nRows after cleaning: {df_clean.shape[0]}")
```

```python
# Unit I: Basic Syntax and Data Types
import pandas as pd
import numpy as np

# Check data types
print("Data Types (Cleaned):\n", df_clean.dtypes)

# Control structure: Verify cleaning
if df_clean[numeric_cols].isnull().sum().sum() == 0:
    print("No missing values in numeric columns!")
else:
    print(f"Remaining missing values: {df_clean[numeric_cols].isnull().sum().sum()}")

# Loop: List columns
print("\nColumns in Dataset:")
for col in df_clean.columns:
    print(col)

# Basic syntax: Shape and sample
print("\nDataset Shape:", df_clean.shape)
print("\nSample Data:\n", df_clean.head())
```

```python
# Unit II: Data Manipulation with NumPy and Pandas
import numpy as np

# NumPy array operations
cropped_array = np.array(df_clean['Cropped land area'])
print("\nNumPy Stats - Cropped land area:")
print("Mean:", np.mean(cropped_array))
print("Std Dev:", np.std(cropped_array))

df_clean['Cropping Intensity'] = df_clean['Cropped land area'] / df_clean['Net sown land area']
print("\nCropping Intensity Sample:\n", df_clean['Cropping Intensity'].head())

# Filter data for more states
selected_states = df_clean[df_clean['State'].isin([
    'Andhra Pradesh', 
    'Assam', 
    'Uttar Pradesh', 
    'Maharashtra', 
    'Tamil Nadu'
])]
print("\nFiltered Data (Selected States):\n", selected_states.head())

# Summary of new column
print("\nCropping Intensity Summary:\n", df_clean['Cropping Intensity'].describe())
```

```python
# Unit III: Data Visualization with Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Matplotlib: Histogram
plt.figure(figsize=(10, 6))
plt.hist(df_clean['Net sown land area'], bins=30, color='lightblue', edgecolor='black')
plt.title('Distribution of Net sown land area', fontsize=14)
plt.xlabel('Net sown land area (hectares)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Seaborn: Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='State', y='Cropped land area', data=df_clean, hue='State', palette='Set2', legend=False)
plt.xticks(rotation=90)
plt.title('Cropped land area by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Cropped land area (hectares)', fontsize=12)
plt.show()

# Seaborn: Pairplot
sns.pairplot(df_clean[numeric_cols + ['Cropping Intensity']].dropna())
plt.show()
```

```python
# Unit IV: Exploratory Data Analysis (EDA)
# Summary statistics
print("Summary Statistics:\n", df_clean[numeric_cols].describe())

# Groupby: State-wise stats
state_summary = df_clean.groupby('State')[numeric_cols].agg(['mean', 'std'])
print("\nTop 5 States by Mean Net sown land area:\n", 
      state_summary['Net sown land area']['mean'].sort_values(ascending=False).head())

# Correlation and covariance
correlation_matrix = df_clean[numeric_cols + ['Cropping Intensity']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
print("\nCovariance Matrix:\n", df_clean[numeric_cols].cov())

# Outlier detection (IQR)
Q1 = df_clean['Net sown land area'].quantile(0.25)
Q3 = df_clean['Net sown land area'].quantile(0.75)
IQR = Q3 - Q1
outliers = df_clean[(df_clean['Net sown land area'] < (Q1 - 1.5 * IQR)) | 
                    (df_clean['Net sown land area'] > (Q3 + 1.5 * IQR))]
print("\nOutliers in Net sown land area:\n", outliers[['District', 'Net sown land area']])

# Visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(y='Net sown land area', data=df_clean)
plt.title('Boxplot of Net sown land area', fontsize=14)
plt.show()
```

```python
# Unit V: Statistical Analysis
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Descriptive statistics
print("Median Forest land area:", df_clean['Forest land area'].median())
print("Mode Cropped land area:", df_clean['Cropped land area'].mode()[0])

# Hypothesis testing: Andhra Pradesh vs. Assam
andhra = df_clean[df_clean['State'] == 'Andhra Pradesh']['Net sown land area']
assam = df_clean[df_clean['State'] == 'Assam']['Net sown land area']
t_stat, p_value = stats.ttest_ind(andhra, assam, equal_var=False)
print(f"\nT-test - T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject null: Significant difference.")

# Probability distribution: Normal fit and Shapiro-Wilk
mu, sigma = stats.norm.fit(df_clean['Net sown land area'])
print(f"\nNormal Fit - Mean: {mu}, Std Dev: {sigma}")
stat, p = stats.shapiro(df_clean['Net sown land area'])
print(f"Shapiro-Wilk - Statistic: {stat}, P-value: {p}")
if p < 0.05:
    print("Not normally distributed.")
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['Net sown land area'], kde=True, stat='density')
x = np.linspace(min(df_clean['Net sown land area']), max(df_clean['Net sown land area']), 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', label='Normal Fit')
plt.title('Net sown land area Distribution', fontsize=14)
plt.legend()
plt.show()

# Variance Inflation Factor (VIF)
X = df_clean[numeric_cols].dropna()
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print("\nVIF Results:\n", vif_data)
```
